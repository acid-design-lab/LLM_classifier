#                              **Классификатор на базе LLM.**

#                                        Инструкция.

1.
Для запуска на локальном ПК, находясь в папке проекта выполните команду
`pip install -r requrements.txt`</br>
Для запуска с сервера пропустите этот шаг.

2. Для запуска скрипта достаточно выполнить команду `python -m classifier <path_to_config_file>` в консоли

3. Образец config файла есть в корневой папке проекта. Скопируйте его и используйте, оставив образец нетронутым. Его можно переименовывать и переносить куда угодно.


#             Содержание файла:
**name=**ivan  # Ваше имя\ник или любой другой идентификатор. Нужно для разделения проводимых экспериментов по папкам. </br></br>
**subject=**chemistry # область в которой будут проводиться ваши эксперименты. Важно указать общее наименование, чтобы не вводить модель в заблуждение. Например: mathematics, physics, marketing и тд.</br></br>
**provider=**openai # название автора использемой модели. На данный момент доступны только openai и mistral. Со временем добавятся и другие LLM.</br></br>
**engine=**gpt-4 # название конкретной сети, которую будете использовать</br></br>
**dataset=**E:\itmo\LLM_fs_classifier\data\text_dataset_random_patterns.csv # Укажите адрес вашего датасета. Требования к датасету будут описаны в следующем пункте.</br></br>
**data_format=**text # в этом пункте нужно указать формат данных для обучения. Всего 2 варианта доступно text или table. Если ваши данные представляют собой сформулированные предложения, то нужно выбрать text, если просто набор признаков, то table</br></br>
**classes=**Cube,Stick,Sphere,Flat,Amorphous # Очень важный пункт. Классификация будет происходить по тем классам, которые тут указаны. Важно указать их с учетом регистра (он должен быть такой же как в датасете). Целевые переменные указываются строго через запятую, без пробелов.</br></br>
**n_for_train=**7 # размер обучающей выборки. Именно столько случайных образцов будет выбрано для передачи контекста модели для классификации.</br></br>
**seed=**36 # нужно для воспроизводимости экспериментов. Фиксируя этот параметр вы будете получать одни и те же ответы модели при одном и том же датасете.</br>


#**требования к датасету**

---
1. Формат файла - .csv или .xlsx</br>
2. заранее удалите из датасета все не имеющие значения колонки, такие как "image_id" или "Unnamed: 0" и тд.</br>
3. В датасете типа table должны остаться только релевантные признаки и целевые переменные. Если это многоклассовая классификация, то целевые переменные должны иметь вид как в pandas.get_dummies()</br>
4. В датасете типа text должна быть только колонка с текстом для классификации и колонки с классами по аналогии pandas.get_dummies()
